### We have to merge all 12 files of the year

```python
# Define the path where your CSV files are stored
path = r'C:\Users\HP\Documents\Sales Data Analysis\Sales_Data' # Update this to the correct path

# List of CSV files to merge
files = ["Sales_January_2019.csv", "Sales_February_2019.csv", "Sales_March_2019.csv", "Sales_April_2019.csv","Sales_May_2019.csv",
        "Sales_June_2019.csv", "Sales_July_2019.csv", "Sales_August_2019.csv", "Sales_September_2019.csv", "Sales_October_2019.csv",
        "Sales_November_2019.csv", "Sales_December_2019.csv"]

# Initialize an empty list to hold the dataframes
dataframes = []

# Read each CSV file into a dataframe and append to the list
for file in files:
    df = pd.read_csv(os.path.join(path, file))
    dataframes.append(df)

# Concatenate all dataframes into a single dataframe
MergedSales2019_df = pd.concat(dataframes, ignore_index=True)

# Save the merged dataframe to a new CSV file
MergedSales2019_df.to_csv(os.path.join(path, 'Merged_Sales_2019.csv'), index=False)

MergedSales2019_df
```

 ## Descriptive Data Analysis with Visualizations

### 1. Quantity Ordered Distribution

```python
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.countplot(data=cleanedMergedSales2019_df, x='Quantity Ordered', palette='viridis')
plt.title('Distribution of Quantity Ordered')
plt.xlabel('Quantity Ordered')
plt.ylabel('Count')
plt.show()
```

### 2. Price Each Distribution

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=cleanedMergedSales2019_df, x='Price Each', bins=50, kde=True, color='blue')
plt.title('Distribution of Price Each')
plt.xlabel('Price Each')
plt.ylabel('Frequency')
plt.show()
```

### 3. Top Products Sold

```python
top_products = cleanedMergedSales2019_df['Product'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_products.values, y=top_products.index, palette='viridis')
plt.title('Top 10 Products Sold')
plt.xlabel('Quantity Sold')
plt.ylabel('Product')
plt.show()
```

### 4. Monthly Sales Trends

```python
cleanedMergedSales2019_df['Month'] = cleanedMergedSales2019_df['Order Date'].dt.month
monthly_sales = cleanedMergedSales2019_df.groupby('Month').sum()['Price Each']
plt.figure(figsize=(10, 6))
sns.lineplot(x=monthly_sales.index, y=monthly_sales.values, marker='o', color='purple')
plt.title('Monthly Sales Trends')
plt.xlabel('Month')
plt.ylabel('Total Sales ($)')
plt.show()
```

## Interactive Dashboard with Plotly

```python
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio

# Set the default renderer to 'notebook' for Jupyter Lab
pio.renderers.default = 'notebook'
```

### 1. Top Products sold

```python
fig = px.bar(top_products, x=top_products.values, y=top_products.index, orientation='h', title='Top 10 Products Sold',
             labels={'x': 'Quantity Sold', 'y': 'Product'})
fig.show()
```

### 2. Monthly Sales Trend

```python
fig = px.line(monthly_sales, x=monthly_sales.index, y=monthly_sales.values, title='Monthly Sales Trends',
              labels={'x': 'Month', 'y': 'Total Sales ($)'})
fig.show()
```

### 3. Sales by Product and Month

```python
sales_by_product_month = cleanedMergedSales2019_df.groupby(['Month', 'Product']).sum()['Price Each'].unstack()
fig = px.imshow(sales_by_product_month, aspect='auto', color_continuous_scale='Viridis', title='Sales by Product and Month')
fig.show()
```

## DASHBOARD

```python
from plotly.subplots import make_subplots

# Create subplots
fig = make_subplots(rows=2, cols=2, subplot_titles=('Quantity Ordered Distribution', 'Price Each Distribution', 
                                                    'Top 10 Products Sold', 'Monthly Sales Trends'))

# Quantity Ordered Distribution
fig.add_trace(go.Histogram(x=cleanedMergedSales2019_df['Quantity Ordered'], nbinsx=10, name='Quantity Ordered'), row=1, col=1)

# Price Each Distribution
fig.add_trace(go.Histogram(x=cleanedMergedSales2019_df['Price Each'], nbinsx=50, name='Price Each'), row=1, col=2)

# Top 10 Products Sold
fig.add_trace(go.Bar(x=top_products.values, y=top_products.index, orientation='h', name='Top 10 Products'), row=2, col=1)

# Monthly Sales Trends
fig.add_trace(go.Scatter(x=monthly_sales.index, y=monthly_sales.values, mode='lines+markers', name='Monthly Sales'), row=2, col=2)

# Update layout
fig.update_layout(height=800, width=1000, title_text="Sales Data Analysis Dashboard", showlegend=False)
fig.show()
```

### Interpretation of the Sales Data Analysis Dashboard

The dashboard is divided into four sections, each providing insights into different aspects of the sales data. Here's a detailed interpretation of each plot:

#### 1. Quantity Ordered Distribution (Top Left)
- **Overview**: This histogram shows the distribution of the quantity ordered.
- **Key Insights**:
  - The majority of orders are for a single unit (1), as indicated by the tall bar at quantity 1.
  - Very few orders are for quantities greater than 1, with a sharp decline in the number of orders as the quantity increases.

#### 2. Price Each Distribution (Top Right)
- **Overview**: This histogram displays the distribution of the prices of the products.
- **Key Insights**:
  - Most products are priced at the lower end of the spectrum, with a significant number of products priced below 100 dollars.
  - There are a few high-priced items, but they are much less common.
  - The distribution is right-skewed, with a long tail towards higher prices.

#### 3. Top 10 Products Sold (Bottom Left)
- **Overview**: This horizontal bar chart lists the top 10 most sold products by quantity.
- **Key Insights**:
  - "USB-C Charging Cable" is the most sold product, followed by "Lightning Charging Cable" and "AAA Batteries (4-pack)".
  - Charging cables and batteries dominate the top-selling products, indicating high demand for these items.
  - Electronic accessories like headphones and monitors are also among the top 10.

#### 4. Monthly Sales Trends (Bottom Right)
- **Overview**: This line plot shows the total sales amount for each month.
- **Key Insights**:
  - There is a clear upward trend in sales as the year progresses, with a notable spike in sales towards the end of the year.
  - The months of May and November show significant peaks, suggesting periods of high sales activity.
  - These peaks could be due to seasonal factors, promotions, or other events driving increased sales.

### Overall Interpretation
- **Sales Volume**: The majority of sales are for single units, indicating that most customers purchase one item per order.
- **Product Pricing**: There is a wide range of product prices, but most sales come from lower-priced items.
- **Popular Products**: Charging cables and batteries are in high demand, reflecting their essential nature and frequent replacement.
- **Seasonal Trends**: Sales increase as the year goes on, with significant spikes in certain months, potentially pointing to seasonal sales trends or successful promotional activities.
